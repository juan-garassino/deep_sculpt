## Step 2: Generate Data 🎨

## Step 3: Train Model 🧠

Train a simple GAN model (this will take a few minutes):

```python
# Quick training with minimal settings
!python main.py train-gan \
    --model-type=simple \
    --epochs=3 \
    --batch-size=4 \
    --void-dim=32 \
    --mixed-precision \
    --output-dir=./quick_results

print("🧠 Training complete!")
```

## Step 4: Generate Samples 🎭

Use your trained model to generate new sculptures:

```python
# Find the trained model
from pathlib import Path
model_files = list(Path('./quick_results').glob('**/generator_final.pt'))

if model_files:
    model_path = model_files[0]
    print(f"🤖 Found model: {model_path}")
    
    # Generate samples
    !python main.py sample-gan \
        --checkpoint={model_path} \
        --num-samples=3 \
        --visualize \
        --output-dir=./quick_samples
    
    print("🎭 Sample generation complete!")
    
    # Show results
    sample_files = list(Path('./quick_samples').glob('*.pt'))
    print(f"📁 Generated {len(sample_files)} samples")
    
    if sample_files:
        import torch
        sample = torch.load(sample_files[0])
        print(f"📏 Sample shape: {sample.shape}")
        print(f"📊 Value range: [{sample.min():.3f}, {sample.max():.3f}]")
        
        print("\n🎉 Success! You've trained your first DeepSculpt model!")
    else:
        print("⚠️ No samples generated")
else:
    print("❌ No trained model found")
```

## Step 5: Visualize Results 👀

Let's see what your model created:

```python
# Try to visualize a generated sample
try:
    from core.visualization.pytorch_visualization import PyTorchVisualizer
    import torch
    
    # Load a sample
    sample_files = list(Path('./quick_samples').glob('*.pt'))
    if sample_files:
        sample = torch.load(sample_files[0])
        
        # Create visualizer
        visualizer = PyTorchVisualizer(backend='plotly')
        
        # Plot the sculpture
        fig = visualizer.plot_sculpture(sample.squeeze())
        if fig:
            fig.show()
            print("🎨 Visualization created!")
        else:
            print("📊 Sculpture statistics:")
            print(f"  Non-zero voxels: {(sample > 0).sum().item()}")
            print(f"  Sparsity: {(sample == 0).float().mean():.3f}")
    else:
        print("❌ No samples to visualize")
        
except Exception as e:
    print(f"⚠️ Visualization error: {e}")
    print("But your model training was successful!")
```
